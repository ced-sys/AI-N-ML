### Synthetic Data generation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(42)
dates=pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
values=np.random.normal(loc=100, scale=20, size=len(dates))
categories=np.random.choice(['A', 'B', 'C', 'D'], size=len(dates))
df=pd.DataFrame({
    'date': dates,
    'value': values,
    'category': categories
})
```

## Common applications of Matplotlib and Seaborn

The provided code demonstrates various visualization techniques used in machine learning and data analysis.

# 1. Line Plot


*   Used for time series analysis, trend detection, and forecasting.
*   The code creates a simple line plot of the first 30 days of data, illustrating fluctuations over time.



```python
plt.figure(figsize=(10, 6))
plt.plot(df['date'][:30], df['value'][:30], marker='o')
plt.title('Basic Line Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show
```

# 2. Scatter plot

*   Useful for analyzing relationships between two numerical variables
*   The scatter plot differentiates data points based on categories, highlighting patterns in different groups.




```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.sample(100), x='date', y='value', hue='category')
plt.title('Scatter Plot by Category')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Box plot


*   Shows data distribution, median, quartiles, and outiers.
*   The Box plot categorizes the data and highlights the presence of outliers.





```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='category', y='value')
plt.title('Box Plot by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

# 4. Histogram with KDE


*   Displays the distribution of a variable and its density estimation.
*   This plots helps in assessing normality and skewness of data.



```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='value', kde=True)
plt.title('Histogram with KDE')
plt.xlabel('Value')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# 5. Heatmap


*   Visualizes correlations or aggregated data in matrix form.
*   The heatmap displays the average values for each category across different months.



```python
pivot_data=pd.pivot_table(
    df,
    values='value',
    index=df['date'].dt.month,
    columns=df['category'],
    aggfunc='mean'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, cmap='YlOrRd', fmt='.1f')
plt.title('Heatmap of Average Values')
plt.xlabel('Category')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

# 6. Pair plot


*   Examines relationships between multiple variables in a dataset
*   Pair plots are useful in feature selection and understanding interactions between variables.



```python
sample_df=df.sample(100).copy()
sample_df['month']=sample_df['date'].dt.month
sns.pairplot(sample_df, hue='category', vars=['value', 'month'])
plt.tight_layout()
plt.show()
```

# 7. Time series with rolling average


*   Used to smooth trends over time, reducing short-term fluctuations.
*   Rolling averages highlights overall trends in time series data.



```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='value', label='Original')
sns.lineplot(
    data=df,
    x='date',
    y=df['value'].rolling(window=30).mean(),
    label='30-day Moving Average'
)

plt.title('Time Series with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# 8. Violin plot


*   Combines box plots and KDE plots to show distributions.
*   Useful in comparing distributions across different categories.



```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='category', y='value')
plt.title('Violin Plot By category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

# 9. Pie chart

*   Represents part-to-whoel relationships.
*   The pie chart visualizaes the distribution of categorical data.



```python
category_counts=df['category'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(category_counts.values,
        labels=category_counts.index,
        autopct='%1.1f%%',
        explode=[0.05]*len(category_counts),
        shadow=True)
plt.title('Distribution of Categories')
plt.axis('equal')
plt.show()
```

##Best Practices and tips

Best Practices and Tips:
1. Always set figure size with plt.figure(figsize=(width, height))
2. Use plt.tight_layout() to prevent label cutoff
3. Rotate x-axis labels when dealing with dates
4. Add proper titles and labels
5. Use appropriate color schemes
6. Consider adding grid lines for better readability
7. Set seed for reproducibility
8. Use appropriate plot types for your data:
   - Line plots: time series data
    - Scatter plots: correlation between variables
    - Box plots: distribution and outliers
    - Heatmaps: correlation matrices or pivot tables
    - Violin plots: distribution comparison
    - Pair plots: relationships between multiple variables
    - Pie charts: part-to-whole relationships
